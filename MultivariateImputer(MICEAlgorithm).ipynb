# Iterative Imputer/MICE (Multivariate Imputer Chained Equations)

# Category of missing data ->
 1. Missing Completely At Random   (Data is missing for no reason at all)
 2. Missing At Random              (Data is missing, but because of another known factor)
 3. Missing Not At Random          (Data is missing because of the value itself)

  MICE ka better result Missing At Random main milega .
 MICE is a method that fills missing values by predicting them using other features again and again.
 Instead of filling missing values once, MICE:fills missing values multiple times using predictions
