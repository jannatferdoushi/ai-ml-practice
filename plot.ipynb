 # Heatmaps

```python
flight_df = sns.load_dataset("flights")
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers") # month has become my index and the second parameter is year it take the position of coln headings and passengers are whole the rest of the data.
flight_df.head()
```

```python
plt.figure (figsize=(12,6))
ax = sns.heatmap(flight_df) # using heatmap function and assing this whole graph to ax variable and printing this ax variable.
ax  # ligther the color bigger the value and darker the color smaller the value.From 1949 to 1960 values are increase.In jan there is lowest number of customise in 1949 but in 1960 yhere is higest number of customise .
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df, annot = True, fmt = "d") # annot = annotation .Shows number inside each box and fmt = d is -> numbers are shown as integers.
ax
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df, annot = True, fmt = "d" , linecolor = "k" , linewidths = "5") # Here k is show black color alag color vi de sakte hain.
ax
                 
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df, annot = True, fmt = "d", linecolor = "k",linewidth="5",cmap = "Blues") # Darker the value larger the number and lighter the value smaller the number .
           
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(flight_df, cbar = False)
```

```python
grid_kws = {"height_ratios":(4,.5), "hspace": .5}
f, (ax, cbar_ax) = plt.subplots(2,gridspec_kw = grid_kws) # Colors shows how big or small the values are -> Dark clr -> lower value ,light clr -> higher value .height_ratios → main heatmap is big, color bar is smalL
ax = sns.heatmap(flight_df, cbar_kws = {"orientation": "horizontal"} , ax = ax,
cbar_ax = cbar_ax)                                                                    # * hspace → space between heatmap and color bar, flight_df → your data
                                                                                      # * cbar_kws={"orientation":"horizontal"} color bar is horizontal,ax=ax → draw heatmap here,cbar_ax=cbar_ax → draw color bar here.
```

```python
plt.figure(figsize=(6,5))
plt.scatter(rollno, marks,color = "green",marker ="*")
plt.show()
```

# Multiple Plots on Same Figure ->

```python
temperature_pune = [25,34,21,45,28,6,43,18,7,2]
humidity_pune  = [28,25,56,34,56,78,34,23,56,23]
temperature_bangalore = [34,56,78,12,34,56,65,43,56,34,21]
humidity_bangalore = [40,35,65,34,78,96,32,55,67,32,21]
```

```python
plt.figure(figsize = (9,9))
plt.plot(temperature_pune , humidity_pune , "ro",ms= 15)
plt.show()
```

```python
# 0,10,20,30,40 are the call xticks and 20,25,30,35,40 are the call yticks
plt.figure(figsize = (8,8))
plt.xticks(np.arange(0,60,5)) # 0 is starting point and 60 is ending point 5 is step
plt.yticks(np.arange(10,60,5))

plt.plot(temperature_pune ,humidity_pune,"ro",ms = 15)
plt.xlabel("temperature")
plt.ylabel("Humidity")
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.xticks(np.arange(0,60,5)) # 0 is starting point and 60 is ending point 5 is step
plt.yticks(np.arange(10,60,5))

plt.plot(temperature_pune ,humidity_pune,"ro",ms = 15)
plt.plot(temperature_bangalore, humidity_bangalore,"bo", ms = 15)
plt.xlabel("temperature")
plt.ylabel("Humidity")
plt.show()
```
