```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv(r"C:\Users\Dr.Sumedha\Downloads\covid_toy.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["has_covid"]),df["has_covid"],test_size=0.2)
```

```python
X_train
```

1. Aam Zindagi ->

```python
# adding simple imputer to fever col 

si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[["fever"]])

# also the test data ->
X_test_fever = si.fit_transform(X_test[["fever"]])

X_train_fever.shape
```

```python
# ordinaLencoding -> cough 
oe = OrdinalEncoder(categories=[["Mild","Strong"]])
X_train_cough = oe.fit_transform(X_train[["cough"]])

# also the test data ->

X_test_cough = oe.fit_transform(X_test[["cough"]])

X_train_cough.shape
```

```python
# OneHotEncoding -> gender,city
ohe = OneHotEncoder(drop="first",sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[["gender","city"]])

# also the test data ->

X_test_gender_city = ohe.fit_transform(X_test[["gender","city"]])

X_train_gender_city.shape
```

```python
# Extracting Age 
X_train_age = X_train.drop(columns=["gender","fever","cough","city"]).values

# also the test data 
X_test_age = X_test.drop(columns=["gender","fever","cough","city"]).values

X_train_age.shape
```

# Mentos Zindagi ->

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[],remainder="passthrough")  # transformer name ka ak object banaya aur thn ColumnTransformer ko cll kiya aur pass karte hain transformer ka list and remainder matlab kavi kavi sab colm ke upar transformer use nehi karna hain jo basha huwa hain uske liya do option hain drop kar dea aur waisha ka waisha hi chor dena.
  
```

```python
transformer = ColumnTransformer(transformers=[
    ("tnf1" , SimpleImputer(),["fever"]),
    ("tnf",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
    ("tnf3",OneHotEncoder(sparse_output=False,drop="first"),["gender","city"])],remainder="passthrough")

 # yeha transformer pass karna hain tuple ke under 
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
