```python
from sklearn.datasets import make_regression  # make_regression() is a function from sklearn.datasets.
# It generates data that follows a linear regression relationship
import numpy as np
```

```python
X,y = make_regression(n_samples = 4, n_features=1,n_informative=1,noise=80,random_state = 13)   
# n_samples = 4 ->This means-Create 4 data points (rows).

# n_features = 1 -> This means-Each sample has 1 input feature (1 column).
# n_informative = 1 -> This means - Out of all features, 1 feature actually affects the output.
# n_informative = 1 -> This means - Out of all features, 1 feature actually affects the output.
# 
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS ->
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red")
```

```python
# Lerts apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0 . ab yehi goal hain ki b ka value 26.15 tak pohchana hain 
y_pred = ((78.35 * X ) + 0).reshape(4)   # This line calculates the predicted values (y_pred) using the equation of a linear regression model.
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="OLS") 
plt.plot(X,y_pred,color="#00a56a",label="b = 0")
plt.legend()
plt.show()
```

Ab gradient descent apply karna hain and red line move towards red line because hame intercept b = 26.15 karna hain.

```python
m = 78.35
b = 0
loss_slope = -2 *np.sum(y - m*X.ravel() - b)   # ravel() converts it into a one-dimensional array. 
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept ->
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="black",label="OLS")
plt.plot(X,y_pred1,color="#00a56a",label="b = {}".format(b)) # this line shows what the regression line looks like after updating b.
plt.plot(X,y_pred,color="#A3E4D7",label="b = 0")
plt.legend()
plt.show()
```
