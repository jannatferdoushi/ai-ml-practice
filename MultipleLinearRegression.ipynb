```python
from sklearn.datasets import make_regression  # Import a function to create fake data,Used for practice (not real data)
import pandas as pd
import numpy as np

import plotly.express as px  # Meaning:Import Plotly (simple graphs),px = shortcut,Used for:Scatter plot,Line graph  and express is used to create It is used to create graphs very easily with simple code.
import plotly.graph_objects as go  # Meaning:Advanced plotting tools, More control over graphs.

from sklearn.metrics import mean_absolute_error,root_mean_squared_error,r2_score
# Import functions to check model performance 
# mean_absolute_error ->Average error,Lower = better
# root_mean_squared_error -> Square error → then root,Punishes big errors more.
# Measures how good model fits data -> Range:1 → perfect ,0 → bad 
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise = 50)
# n_samples=100 -> Number of rows (data points),Means:100 data entries
# n_features=2 -> Number of input columns,Means:2 inputs (like x1, x2)

# n_informative=2 -> Important features that affect output, Means: Both features are useful, Example:x1 affects y ,x2 affects y ✅

# n_targets=1 ->  Number of output columns, Means: Only one output (y)

# noise=50 -> Adds randomness (error)n , Means:,Data is not perfectly straight line,More realistic.
```

```python
df = pd.DataFrame({"feature1":X[:,0],"feature2":X[:,1],"target":y})
# pd.DataFrame -> creates a table.
# "feature2": X[:,1] -> Means:Take all rows,From column 1,1 = second column
# "target": y -> This is  output column.
```

```python
df.shape
```

```python
df.head()
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')  # This line creates a 3D scatter plot (graph) using my DataFrame

fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
 # import LinearRegression ->  This class is used to:Train model, Find m and b, Predict values.
```

```python
lr = LinearRegression() # creating a Linear Regression model object.
```

```python
lr.fit(X_train,y_train)  # training the Linear Regression model .
# It already knows:

# slope (m)
# intercept (b)
```

```python
y_pred = lr.predict(X_test) # 
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

xGrid, yGrid = np.meshgrid(x, y)

# FIRST create final
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T

# THEN predict
z_final = lr.predict(final).reshape(10, 10)

z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.add_trace(go.Surface(x = x, y = y, z =z ))

fig.show()
     
```
