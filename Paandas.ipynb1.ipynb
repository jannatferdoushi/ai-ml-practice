# Filling the Null Values 

```python
df = pd.read_csv(r"C:\Users\Dr.Sumedha\Desktop\New folder (2)\sample.csv")
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum() # Give number of null values each and every coln individually.
```

```python
df.fillna(2)  # Jo vi null value hain woh 0 ya 2 ya kuch aur se fill kar sakte hain .
```

```python
df.fillna({"Physics" : "none","Chemistry" : 0,"Maths" :30,"Computer" : "!"})    # We can fill different different thing
```

```python
df.fillna(method = "ffill",axis = 1) # All the null value will be filled in the just previous value . axis  = 1 likhne se jo null value hain woh fill hota hain just previous coln se.
```

```python
df["Physics"].fillna(value = df["Physics"].mean()) # I am going to find the avg marks of physics of complete class and i am going to fill those marks at the null values .45 which is avg of all this marks so we can fill the null value with the mean.
```

```python
df.fillna(method = "bfill",inplace  = True) # backward fill -> fill the null value with  the next value not previous . If provide inplace  = True then it is going to replace the original DataFrame.
df
```

# Merge()Function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[34,67,34,89,12]}) 
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[78,33,39,81,90]}) 
df2
```

```python
pd.merge(df1, df2, on = 'Roll No.')
```

```python
pd.merge(df2, df1, on = 'Roll No.')
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2) # if Roll no. is a intersecting coln if i have intersecting coln thn donot need to provide on parameter .
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]}) # when run  
df4 = pd.DataFrame({'Roll No.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})                             
```

```python
pd.merge(df3,df4) # when run pd.merge(df3,df4) it is going to take only those roll numbers who are intersecting.1,2,3 roll numbers are intersecting.
```

```python
pd.merge(df3,df4,how = 'left')
```

```python
pd.merge(df3,df4,how = 'right')
```

```python
pd.merge(df3,df4,how = 'outer')
```
