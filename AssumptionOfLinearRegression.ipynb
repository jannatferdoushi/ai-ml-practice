```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Dr.Sumedha\Downloads\data.csv")
df.head()
```

```python
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()  # This line creates an object of the LinearRegression class.and The new object is stored in the variable model.
    
model.fit(X_train,y_train)  # Train the model using the training data.
```

```python
# Residual    # It is written to tell the reader that the next lines calculate the residuals (errors)
y_pred = model.predict(X_test)   # Use the trained model to predict the output for new input data
residual = y_test - y_pred   # This calculates the residual (prediction error).
```

# 1. Linear Relationship ->  

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 2.5)) # ncols3= Creates 3 graphs in one row.

ax1.scatter(df['feature1'], df['target'])
ax1.set_title("Feature1")
ax2.scatter(df['feature2'], df['target'])
ax2.set_title("Feature2")
ax3.scatter(df['feature3'], df['target'])
ax3.set_title("Feature3")

plt.show()
```

 # 2. Multicollinearity  ->

```python
# This code is used to calculate the Variance Inflation Factor (VIF) for every feature in your dataset.
from statsmodels.stats.outliers_influence import variance_inflation_factor # It imports the variance_inflation_factor() function from the statsmodels library.

vif = []   # vif tells How much one feature can be explained by the other features. vif=[]  -> This creates an empty list.

for i in range(X_train.shape[1]):   
    # shape[0] → Number of rows

    # shape[1] → Number of columns
    vif.append(variance_inflation_factor(X_train, i)) # This function calculates the VIF for the feature at index i.
```

```python
pd.DataFrame({'vif': vif}, index=df.columns[0:3]).T
```
