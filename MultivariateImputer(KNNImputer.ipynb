```python
import pandas as pd 
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # Scikit-learn (sklearn) is a Python library used to:Build machine learning models,Train models,Test and evaluate models
# sklearn → Refers to the library Scikit-learn
#metrics → A module that contains functions to evaluate models
# accuracy_score → A function used to measure how correct your model predictions
```

```python
df = pd.read_csv(r"C:\Users\Dr.Sumedha\Downloads\train (2).csv")[["Age","Pclass","Fare","Survived"]]
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=["Survived"])
y = df["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
knn = KNNImputer()

X_train_trf = knn.fit_transform(X_train)  # X_train_trf = transformed version of X_train (after filling missing values)
X_test_trf = knn.fit_transform(X_test)    #  X_test_trf is the transformed version of your test data (X_test) after handling missing values.
# fit should be done only on training data
# Test data should NOT be used for learning 
```

```python
pd.DataFrame(X_train_trf,columns=X_train.columns)  # This line converts transformed data (X_train_trf) back into a DataFrame with proper column names.
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf, y_train)  # The model learns relationship between input and output

y_pred = lr.predict(X_test_trf) # Model gives predicted answers for test data- Output stored in y_pred

accuracy_score(y_test, y_pred) # Compare actual vs predicted and calculate accuracy
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
