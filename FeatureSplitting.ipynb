```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score  # It is used to check how good your model is , Instead of testing model once ,It tests model multiple times on different data parts ✔


from sklearn.linear_model import LogisticRegression  # It is a machine learning model used for classification,It predicts YES/NO, 0/1, Pass/Fail
```

```python
df=pd.read_csv(r"C:\Users\Dr.Sumedha\Downloads\train (2).csv")[["Age","Pclass","SibSp","Parch","Survived"]]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# iloc is used to select data by position (index number)
X = df.iloc[:,0:4]  # Select all rows and columns from index 0 to 3
y = df.iloc[:,-1]  # -1 matlab Select last column and : matlab all rows 
```

```python
df.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring="accuracy",cv=20))  # Average accuracy of model using 20-fold cross-validation
```

```python
X["Family_size"] = X["SibSp"] + X["Parch"] + 1 # Add 1 for the person himself/herself (People who are traveling together with that person)
```

```python
X.head()
```

```python
def myfunc(num):   # ak function banaya aur family size ko provide kiya 
    if num == 1:
        # alone 
        return 0

    elif  num > 1 and num <= 4:
        # small family
        return 1

    else :  # large family
        return 2
        
    
```

```python
myfunc(4)
```

```python
X["Family_type"] = X["Family_size"].apply(myfunc)
X.head()
```

```python
X.drop(columns=["SibSp","Parch","Family_size"],inplace=True)
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring="accuracy",cv=20))
```
