```python
df.isnull().mean()*100
```

```python
X = df.drop(columns = ["Survived"])  # df = your dataset (table),So X contains all columns except Survived
y = df["Survived"]  # Selects only the Survived column,So y contains:0 → did not survive,1 → survived
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)  # Splits X and y,Creates train & test datasets,Uses 20% data for testing,Keeps result same using random_state=2
```

```python
X_train["Age_imputed"] = X_train["Age"]  # Takes the "Age" column from X_train,Creates a new column "Age_imputed",Copies all values into it.
X_test["Age_imputed"] = X_test["Age"]
```
