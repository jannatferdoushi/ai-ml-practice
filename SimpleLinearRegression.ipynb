# Simple Linear Regression ->
Simple Linear Regression is a method that uses one input variable to predict an output using a straight-line equation.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Dr.Sumedha\Downloads\placement (2).csv")
df.head()
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("cgpa")     # Names the horizontal axis - Now X-axis shows → “cgpa”
plt.ylabel("package(in lpa)")  # Names the vertical axis-“lpa” = lakh per annum
```

```python
x = df.iloc[:,0:1]   # input colm 
y = df.iloc[:,-1]     # target colm
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()  # storing that model in a variable named lr 
```

```python
lr.fit(x_train,y_train)  # lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))  # .values-Converts pandas data → NumPy array  and  . reshape(1,1) means Model expects input in 2D format
# Take first test value
# Convert to correct format
# Send to model
# Get prediction
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.plot(x_train,lr.predict(x_train),color="red")  # train main thora lamba hoga line test se
plt.xlabel("cgpa")     
plt.ylabel("package(in lpa)")  
```

```python
m = lr.coef_   # lr.coef_ gives the slope (weight) of the model
```

```python
m
```

```python
b = lr.intercept_
```
